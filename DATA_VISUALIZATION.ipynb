```python
features_mean=list(df.columns[0:10])
# split dataframe into two based on diagnosis
dfM=df[df['HeartDisease'] ==1]
dfB=df[df['HeartDisease'] ==0]
```

```python
#Stack the data
plt.rcParams.update({'font.size': 8})
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(8,10))
axes = axes.ravel()
for idx,ax in enumerate(axes):
    ax.figure
    binwidth= (max(df_heart_2[features_mean[idx]]) - min(df_heart_2[features_mean[idx]]))/50
    ax.hist([dfM[features_mean[idx]],dfB[features_mean[idx]]], bins=np.arange(min(df_heart[features_mean[idx]]), max(df_heart[features_mean[idx]]) + binwidth, binwidth) , alpha=0.5,stacked=True, label=['positive','negative'],color=['r','g'])
    ax.legend(loc='upper right')
    ax.set_title(features_mean[idx])
plt.tight_layout()
plt.show()
```
